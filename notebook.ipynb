```python
import os
import numpy as np

# Create a directory to store the preprocessed labels
preprocessed_label_dir = './data/data_object_label_2_preprocessed/'
os.makedirs(preprocessed_label_dir, exist_ok=True)

# Define the class map (KITTI object classes)
class_map = {
    'Car': 1,
    'Pedestrian': 2,
    'Cyclist': 3,
    'Truck': 4,
    'Van': 5,
    'Tram': 6,
    'Person_sitting': 7,
    'Misc': 8,
    'DontCare': 9  # We will ignore these during training
}

# Loop through all label files, parse them, and save them as .npy files
for idx, label_file in enumerate(label_files):
    # Construct the full path to the label file
    label_path = os.path.join(label_dir, label_file)
    
    # Initialize lists to store bounding boxes and class labels
    boxes = []
    class_indices = []

    # Open and read the label file
    with open(label_path, 'r') as f:
        for line in f.readlines():
            parts = line.strip().split()  # Split the line into parts
            
            # Extract the class name
            class_name = parts[0]
            
            # Skip 'DontCare' objects (not needed for training)
            if class_name == 'DontCare':
                continue

            # Extract bounding box coordinates (xmin, ymin, xmax, ymax)
            xmin = float(parts[4])
            ymin = float(parts[5])
            xmax = float(parts[6])
            ymax = float(parts[7])
            
            # Append the bounding box and class index
            boxes.append([xmin, ymin, xmax, ymax])
            class_indices.append(class_map[class_name])  # Map class name to index

    # Save the parsed labels as a dictionary in a .npy file
    label_data = {'boxes': np.array(boxes, dtype=np.float32), 'labels': np.array(class_indices, dtype=np.int64)}
    np.save(os.path.join(preprocessed_label_dir, f'label_{idx}.npy'), label_data)

    # Print progress every 100 labels
    if idx % 100 == 0:
        print(f"Processed {idx} labels...")

print(f"Preprocessed and saved {len(label_files)} labels.")
```

# KittiDataset class     

**Import libraries**

```python
import os
import numpy as np
import torch
from torch.utils.data import Dataset
```

**Declaration of the class**

```python
class KittiDataset(Dataset):
    def __init__(self, image_dir, label_dir):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.image_files = sorted(os.listdir(image_dir))
        self.label_files = sorted(os.listdir(label_dir))

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        """
        Retrieve the image and corresponding label by index.
        """
        # Load the preprocessed image
        image_path = os.path.join(self.image_dir, self.image_files[idx])
        image = np.load(image_path)  
        image = torch.tensor(image).permute(2, 0, 1)  # Convert to [C, H, W] for tensor form input to model

        # Load the label and the target dictionary expected by Faster R-CNN
        label_path = os.path.join(self.label_dir, self.label_files[idx])
        label_data = np.load(label_path, allow_pickle=True).item()

        target = {
            'boxes': torch.tensor(label_data['boxes'], dtype=torch.float32),
            'labels': torch.tensor(label_data['labels'], dtype=torch.int64)
        }

        return image, target
```

**Integrate with Dataloader**

```python
from torch.utils.data import DataLoader

# Define a custom collate function to handle variable-sized tensors
def collate_fn(batch):
    images, targets = zip(*batch)  # Unzip the batch

    # Stack images into a single tensor
    images = torch.stack(images, dim=0)

    # Keep targets as a list (since each target may have a different size)
    targets = list(targets)

    return images, targets

# Define paths to preprocessed image and label directories
right_image_dir = './data/data_object_image_2_preprocessed/'
label_dir = './data/data_object_label_2_preprocessed/'

# Initialize the Dataset
dataset = KittiDataset(right_image_dir, label_dir)

# Initialize the DataLoader with custom collate_fn
dataloader = DataLoader(dataset, batch_size=4, shuffle=True, num_workers=0, collate_fn=collate_fn)

# Function to load and print batches with error handling
def test_dataloader(dataloader):
    batch_idx = 0  # Initialize batch_idx to avoid UnboundLocalError
    try:
        for batch_idx, (images, targets) in enumerate(dataloader):
            print(f"Batch {batch_idx + 1} loaded successfully!")
            print(f"  - Image batch shape: {images.shape}")
            for i, target in enumerate(targets):
                print(f"    - Target {i}: Boxes - {target['boxes'].shape}, Labels - {target['labels'].shape}")
    except Exception as e:
        print(f"Error occurred while loading batch {batch_idx + 1}: {str(e)}")

# Call the function to test DataLoader
print("Starting DataLoader test...")
test_dataloader(dataloader)
print("DataLoader test completed.")
```

```python
# Fetch one batch of data
images, targets = next(iter(dataloader))

# 1. Check the shape of the images
print(f"Batch of images shape: {images.shape}")  # Expected: [batch_size, channels, height, width]

# 2. Check the content and shape of the targets
for i, target in enumerate(targets):
    print(f"\nTarget for Image {i}:")
    print(f"  - Boxes shape: {target['boxes'].shape}")  # Expected: [num_boxes, 4]
    print(f"  - Labels shape: {target['labels'].shape}")  # Expected: [num_boxes]

    # Optional: Print the first box and label to inspect
    print(f"  - First Box: {target['boxes'][0]}")
    print(f"  - First Label: {target['labels'][0]}")

# 3. Verify data types (Images and Targets)
print(f"\nImage Tensor Type: {images.dtype}")  # Should be torch.float32
print(f"Target Box Tensor Type: {targets[0]['boxes'].dtype}")  # Should be torch.float32
print(f"Target Label Tensor Type: {targets[0]['labels'].dtype}")  # Should be torch.int64
```

**Setting up the model**

```python
import torch
from torchvision.models.detection import fasterrcnn_resnet50_fpn

# Load the pre-trained Faster R-CNN model
model = fasterrcnn_resnet50_fpn(pretrained=True)

# Modify the classifier head for KITTI (9 classes + 1 background = 10 classes)
num_classes = 10  

# Get the number of input features for the classifier and box regressor
in_features = model.roi_heads.box_predictor.cls_score.in_features
in_box_features = model.roi_heads.box_predictor.bbox_pred.in_features

# Replace the classifier head
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

# Move the model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

print(f"Model is ready and moved to: {device}")
```

**Setting Up the Optimizer**

```python
import torch.optim as optim

# Define the optimizer
optimizer = optim.SGD(
    model.parameters(), 
    lr=0.005, 
    momentum=0.9, 
    weight_decay=0.0005
)
```

**Training Loop**

```python
# Training loop parameters
num_epochs = 5  # You can increase this for better results
model.train()  # Set the model to training mode

for epoch in range(num_epochs):
    epoch_loss = 0  # Track loss for each epoch

    print(f"\nEpoch {epoch + 1}/{num_epochs}")
    for batch_idx, (images, targets) in enumerate(dataloader):
        # Move images and targets to the same device as the model
        images = [img.to(device) for img in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        # Zero the optimizer gradients
        optimizer.zero_grad()

        # Forward pass: compute the loss
        loss_dict = model(images, targets)
        total_loss = sum(loss for loss in loss_dict.values())

        # Backward pass: compute gradients
        total_loss.backward()

        # Update weights
        optimizer.step()

        # Accumulate loss for reporting
        epoch_loss += total_loss.item()

        # Print progress every 10 batches
        if batch_idx % 10 == 0:
            print(f"  Batch {batch_idx + 1}, Loss: {total_loss.item():.4f}")

    print(f"Epoch {epoch + 1} completed. Total Loss: {epoch_loss:.4f}")

print("Training completed!")
```

**Save The Model**

```python
# Save the trained model's state_dict
torch.save(model.state_dict(), 'faster_rcnn_kitti.pth')
print("Model saved successfully!")
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import os

# Define the path to the test images
test_image_dir = './data/data_object_image_2/testing/image_2/'

# Function to visualize predictions on the image
def visualize_predictions(image, predictions, class_names):
    """
    Display image with predicted bounding boxes and class labels.
    Args:
    - image: Input image tensor of shape [C, H, W]
    - predictions: Model's output containing 'boxes' and 'labels'
    - class_names: List of class names corresponding to the labels
    """
    # Move the image to CPU and convert to NumPy format
    image = image.cpu().permute(1, 2, 0).numpy()

    # Plot the image
    plt.imshow(image)
    ax = plt.gca()

    # Draw bounding boxes and class labels
    for box, label in zip(predictions['boxes'], predictions['labels']):
        # Move tensors to CPU and extract values
        box = box.cpu().numpy() 
        label = label.cpu().item()

        # Extract box coordinates (xmin, ymin, xmax, ymax)
        xmin, ymin, xmax, ymax = box

        # Draw a rectangle around the detected object
        rect = plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                             fill=False, color='red', linewidth=2)
        ax.add_patch(rect)

        # Add the class label text near the top-left corner of the box
        ax.text(xmin, ymin, class_names[label], color='blue', fontsize=12)

    plt.axis('off')
    plt.show()

# List all test images
test_image_files = sorted(os.listdir(test_image_dir))

# Select a test image for inference
test_image_path = os.path.join(test_image_dir, test_image_files[0])

# Load the test image using OpenCV and convert it to a PyTorch tensor
image = plt.imread(test_image_path)  # Or use cv2.imread() if OpenCV is preferred
image = torch.tensor(image, dtype=torch.float32).permute(2, 0, 1) / 255.0  # Normalize
image = image.unsqueeze(0).to(device)  # Add batch dimension and move to GPU/CPU

# Set the model to evaluation mode and perform inference
model.eval()
with torch.no_grad():
    predictions = model(image)[0]  # Get the first result from the batch

# Define the KITTI class names (for the labels)
class_names = [
    "Background", "Car", "Pedestrian", "Cyclist", "Truck", 
    "Van", "Tram", "Person_sitting", "Misc", "DontCare"
]

# Visualize the predictions on the selected test image
visualize_predictions(image[0], predictions, class_names)
```
